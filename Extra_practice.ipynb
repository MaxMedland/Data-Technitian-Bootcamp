```python
def userInput(): # recursive
  try: # every input is formated to request an int
    u = int(input())
  except:
    print("Invalid input. Please try again")
    userInput()
  return u

def checkPIN():
  print("Please enter your PIN")
  pin = userInput()
  if pin == 1234: # hardcoded pin number
    return True
  else:
    return False

def mainMenu():
  print("Menu\n1-Display current ballance\n2-Make a withdrawal\n3-Make a deposit\n4-Exit")
  u = userInput()
  return u

def displayBalance(b):
  print(f"Your current balance is: £{b}")

def makeWithdrawal(b): # recursive
  print("How much would you like to withdraw?\n1-£5\n2-£10\n3-£20\n4-£50\n5-Other ammount")
  u = userInput()
  if u == 1:
    v = 5
  elif u == 2:
    v = 10
  elif u == 3:
    v = 20
  elif u == 4:
    v = 5
  else:
    print("Invalid input. Please try again")
    makeWithdrawal(b)
  b -= v
  print(f"You withdrew: £{v}")
  return b

def makeDeposit(b):
  print("How much would you like to deposit?")
  u = userInput()
  b += u
  print(f"You deposited: £{u}")
  return b

def ATM():
  entry = checkPIN()
  if entry == False:
    print("Invalid PIN.")
    return
  u = 0
  balance = 10000 # random ammount hardcoded to function
  while u != 4:
    u = mainMenu()
    if u == 1:
      displayBalance(balance)
    elif u == 2:
      balance =  makeWithdrawal(balance)
    elif u == 3:
      balance = makeDeposit(balance)
    elif u == 4:
      print("Thank you for using this ATM")
      return
    else:
      print("Invalid input. Please try again")

ATM()
```
